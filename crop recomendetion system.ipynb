**Data Preprocessing**

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python

categorical_columns = ["label"]
le = LabelEncoder()
for col in categorical_columns:
    df[col] = le.fit_transform(df[col])
df.head()
```

**To get the actual value from label encoder value**

```python
label_mapping = dict(zip(le.classes_, le.transform(le.classes_)))
reverse_mapping = {v: k for k, v in label_mapping.items()}
print("Label Mapping (Crop to Numerical):", label_mapping)
```

**split the data set**

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

**To balance the data set **//normalization feature in a dataset

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X_scaled, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

**Apply this algorithms to the dataset for traning**

```python

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
import xgboost as xgb
import catboost as cb
from sklearn.metrics import accuracy_score
#from sklearn.metrics import precision_score, recall_score


# Initialize models
models = {
    "Logistic Regression": LogisticRegression(max_iter=10000),
    "Decision Trees": DecisionTreeClassifier(),
    "Random Forests": RandomForestClassifier(),
    "Gradient Boosting Machines (GBM)": GradientBoostingClassifier(),
    "Support Vector Machines (SVM)": SVC(),
    "k-Nearest Neighbors (k-NN)": KNeighborsClassifier(),
    "Naive Bayes": GaussianNB(),
    "AdaBoost": AdaBoostClassifier(),
    "XGBoost": xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss'),
    "CatBoost": cb.CatBoostClassifier(verbose=0)
}

results = {}

# Train and evaluate models
for name, model in models.items():
    # Train the model
    model.fit(X_train, Y_train)

    # Predict and evaluate on training data
    Y_train_pred = model.predict(X_train)
    train_accuracy = accuracy_score(Y_train, Y_train_pred)

    # Predict and evaluate on testing data
    Y_test_pred = model.predict(X_test)
    test_accuracy = accuracy_score(Y_test, Y_test_pred)
    #precision = precision_score(Y_test, Y_test_pred)
    #recall = recall_score(Y_test, Y_test_pred)

    # Store results
    results[name] = {
        'Training Accuracy': train_accuracy,
        'Testing Accuracy': test_accuracy,
        #'Precision': precision,
        #'Recall': recall
    }

    # Print the results
    print(f"{name}:")
    print(f"  Training Accuracy: {train_accuracy:.4f}")
    print(f"  Testing Accuracy: {test_accuracy:.4f}")
    #print(f"  Precision: {precision:.4f}")
    #print(f"  Recall: {recall:.4f}\n")

# Find the best model based on testing accuracy
best_model_name = max(results, key=lambda x: results[x]['Testing Accuracy'])
best_model_accuracy = results[best_model_name]['Testing Accuracy']
best_model = models[best_model_name]

print(f"The best model based on testing accuracy is: {best_model_name} with a Testing Accuracy of {best_model_accuracy:.4f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Prepare the data for the bar plot
model_names = list(results.keys())
train_accuracies = [results[model]['Training Accuracy'] for model in model_names]

# Set the positions for the bars
x = np.arange(len(model_names))  # The label locations
width = 0.5  # The width of the bars

# Define a simple list of unique colors for the bars (you can modify these colors)
train_colors = ['red', 'green', 'blue', 'cyan', 'magenta', 'yellow', 'orange', 'purple', 'brown', 'pink']

# Create the figure and axis for the training accuracy bar chart
fig, ax = plt.subplots(figsize=(10, 6))

# Create bars for training accuracy with unique colors
bars_train = ax.bar(x, train_accuracies, width, color=train_colors[:len(model_names)])

# Add annotations to display accuracy values on the bars
for bar in bars_train:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.02, f'{yval:.4f}', ha='center', va='bottom', fontsize=10, color='black')

# Customize the appearance
ax.set_xlabel('Machine Learning Models', fontsize=12)
ax.set_ylabel('Training Accuracy', fontsize=12)
#ax.set_title('Training Accuracy of Different Models', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(model_names, rotation=45, ha='right')
ax.set_ylim(0, 1)  # Set y-axis limit from 0 to 1 since accuracy is between these values

# Show the plot
plt.tight_layout()
plt.show()
```
